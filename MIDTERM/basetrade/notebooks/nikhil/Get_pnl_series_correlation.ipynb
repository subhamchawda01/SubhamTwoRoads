```python
import os
import sys
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import shutil

sys.path.append(os.path.expanduser('~/basetrade/')) 
from walkforward.definitions import execs
from walkforward.utils import date_utils
from walkforward.utils.run_exec import exec_function
```

```python
def load_config_pnl_for_strat(work_dir, strat_name, start_date , end_date, config_date_pnl_test):
    summarize_pnl_cmd = execs.execs().summarize_local_results_dir_and_choose_by_algo + " kCNAPnlSharpe\
    4000 4000 -1 1 5000 100000 " +work_dir + "/local_results_base_dir/ " + str(start_date) + " " + str(end_date)

    PnL_data = exec_function(summarize_pnl_cmd)[0].strip()
    PnL_data = PnL_data.split("\n")
    #print PnL_data
    for line in PnL_data:
        if line == "" : continue
        words = line.split(" ")
        if words[0] == "STRATEGYFILEBASE":
            config = str(words[1])

            config_date_pnl_test[config] = {}
            continue
        if words[0] == "STATISTICS":
            continue
        if config != strat_name:
            #print config
            continue            
        #print config
        date_ = int(words[0])
        config_date_pnl_test[config][date_] = words[1]
```

```python
def load_config_date_pnl_map(shc , p_start_time , p_end_time , start_date , end_date, config_date_pnl):

    #Get the PnL data for the pool using summarize_strategy
    summarize_pnl_dir = execs.execs().modelling + "/wf_strats/" + shc + "/" + p_start_time + "-" + p_end_time + "/"
    summarize_pnl_pool = p_start_time + "-" + p_end_time
    summarize_pnl_command = [execs.execs().summarize_strategy , shc ,summarize_pnl_pool , "DB" , str(start_date) , end_date , "IF" , "kCNAPnlSharpe" , "0" , "IF" , "0"]
    summarize_pnl_command = " ".join(summarize_pnl_command)
    #print summarize_pnl_command
    PnL_data = exec_function(summarize_pnl_command)[0].strip()

    #fill the map of config_date_pnl from output of summarize strategy results
    PnL_data = PnL_data.split("\n")
    for line in PnL_data:
        if line == "" : continue
        words = line.split(" ")
        if words[0] == "STRATEGYFILEBASE":
            config = words[1]
            config_date_pnl[config] = {}
            continue
        if words[0] == "STATISTICS":
            continue
        date_ = int(words[0])
        config_date_pnl[config][date_] = words[1]
        
```

```python
config_date_pnl = {}
config_date_pnl_test = {}
config_date_pnl_test_2 = {}
shc = 'USD000UTSTOM'
p_start_time = 'MSK_1010'
p_end_time =  'MSK_1400'
start_date = "20160701"
end_date = "20170704"
work_dir = "/media/shared/ephemeral16/dvctrader/pnl_modelling/USD000UTSTOM/1499345432217/"
work_dir = "/media/shared/ephemeral16/dvctrader/pnl_modelling/USD000UTSTOM/1500383702323/"


work_dir_2 = "/media/shared/ephemeral16/dvctrader/pnl_modelling/USD000UTSTOM/1500285718814"
strat_name = "strat_452_0"
strat_name = "strat_3_0"
strat_name_2 = "strat_58_0"

load_config_date_pnl_map(shc , p_start_time , p_end_time , start_date , end_date, config_date_pnl)
load_config_pnl_for_strat(work_dir, strat_name, start_date , end_date, config_date_pnl_test)
load_config_pnl_for_strat(work_dir_2, strat_name_2, start_date , end_date, config_date_pnl_test_2)
```

```python
configs_printed = []
corrs = []
for config in [strat_name]:
    #print config
    for config2 in config_date_pnl:
        if config != config2 and [config,config2] not in configs_printed :
            
            pnl_1 = []
            pnl_2 = []
            for date in config_date_pnl_test[config]:
                if date in config_date_pnl[config2] and date in config_date_pnl_test[config]:
                    pnl_2.append(int(config_date_pnl[config2][date]))
                    pnl_1.append(int(config_date_pnl_test[config][date]))
    
            configs_printed.append(([config,config2] ))
            configs_printed.append([config2,config])
            np_pnl_1 = np.array(pnl_1)
            np_pnl_2 = np.array(pnl_2)
            np_pnl_3 = np_pnl_1 + np_pnl_2
            print config,config2,np_pnl_1.mean()/np_pnl_1.std(),np_pnl_2.mean()/np_pnl_2.std(),np.corrcoef(pnl_1,pnl_2)[0][1], len(pnl_1)#,np_pnl_3.mean(),np_pnl_3.mean()/np_pnl_3.std() 
            corrs.append(np.corrcoef(pnl_1,pnl_2)[0][1])
np_corrs  = np.array(corrs)
print np_corrs.max()
```
