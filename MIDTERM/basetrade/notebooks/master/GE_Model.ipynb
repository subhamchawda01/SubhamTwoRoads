## GE Curve based Models 

### Risk_adusted Models
In most of the models we learn in in data space, we maximize the pnl-contribution irrespective of the risk involved
Here we would try to incorporate the overall risk of the portfolio as well in learning model for each securities. 

#### Formulation

With assumption that we get filled for each order we place, following would be position for each security.

Let's assume model value for each expiry is m_i, stdev is sigma_i

risk(i) = m_i/ sigma_i
position(i) = m_i/(sigma_i)^2

if the future change in price is y_i then
pnl(i) = position(i)* y_i

cost function = maximize(pnl(i) - lambda * position(i))

```python
def get_expiry_contract_from_column_name(column_name, expiry_index):
    # examples of column_name 'V_GE_0'
    column_words = column_name.split('_')
    current_expiry = column_words[-1]
    new_expiry = int(current_expiry) + expiry_index
    if new_expiry < 0 or new_expiry >15 :
        return ''
    else:
        return column_words[-2] + '_' + str(new_expiry)
```

```python
for name in signal.columns:
    # for each expiry compute the model prediction separately
    trend_column = name.replace('P', 'T')
    last_maturity = get_expiry_contract_from_column_name(trend_column, -1)
    next_maturity = get_expiry_contract_from_column_name(trend_column, +1)

    if last_maturity and next_maturity:
        signal[name] = (trend['T_' + last_maturity] + trend['T_' + next_maturity])*0.5
    else:
        # should only happen in case of 'GE_0' and 'GE_15'
        print ('Using Self Trend for ' + trend_column)
        signal[name] = trend[trend_column]
    #signal[name] = trend[trend_column] 
```
